```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
data = pd.DataFrame({
    'A': np.random.randn(100),
    'B': np.random.randint(1, 5, 100),
    'C': np.random.uniform(0, 1, 100)
})

# Summary statistics
print("Summary Statistics:")
print(data.describe())

# Data visualization
# Histogram
plt.figure(figsize=(10, 6))
data['A'].plot(kind='hist', bins=20, color='skyblue', alpha=0.7)
plt.title('Histogram of column A')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Boxplot
plt.figure(figsize=(10, 6))
data.boxplot(column='A', grid=False)
plt.title('Boxplot of column A')
plt.ylabel('Values')
plt.show()

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['B'], data['C'], color='green', alpha=0.7)
plt.title('Scatter Plot of columns B and C')
plt.xlabel('B')
plt.ylabel('C')
plt.grid(True)
plt.show()
```
